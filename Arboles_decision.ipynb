```python
# Dependencias
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer, load_diabetes
from sklearn.inspection import DecisionBoundaryDisplay
from scipy.stats import multivariate_normal
from matplotlib import pylab as plt
import pandas as pd
import seaborn as sns
```

```python
# Los tres grupos de clasificación
seed = 1
X_1 = multivariate_normal(mean=[5, 5],
                          seed=seed,
                          cov=[[4, 0], [0, 2]]).rvs(1000)
X_2 = multivariate_normal(mean=[-5, -10],
                          seed=seed,
                          cov=[[2, 1], [1, 3]]).rvs(1000)
X_3 = multivariate_normal(mean=[15, -6],
                          seed=seed,
                          cov=[[2, 3], [3, 7]]).rvs(1000)
df = pd.DataFrame([dict(x=x, y=y, clase=1) for x, y in X_1] + \
                  [dict(x=x, y=y, clase=2) for x, y in X_2] + \
                  [dict(x=x, y=y, clase=3) for x, y in X_3])
sns.relplot(data=df, kind='scatter',
            x='x', y='y', hue='clase')
```

```python
X = np.concatenate((X_1, X_2, X_3), axis=0) # Hacemos el conjunto completo
y = np.array([1] * 1000 + [2] * 1000 + [3] * 1000) # Etiquetas
arbol = tree.DecisionTreeClassifier(criterion='entropy').fit(X, y)
_ = tree.plot_tree(arbol, node_ids=True,
                   feature_names=['x', 'y'], label='root')
```

```python
# Eliminando la hoja 6: 
ax = plt.subplot(2, 1, 1)
arbol = tree.DecisionTreeClassifier(criterion='entropy', min_samples_split=1003).fit(X, y)
_ = tree.plot_tree(arbol, node_ids=True,
                   feature_names=['x', 'y'], label='root')
ax = plt.subplot(2, 1, 2)
DecisionBoundaryDisplay.from_estimator(arbol, X, cmap=plt.cm.RdYlBu,
                                       response_method='predict',
                                       ax=ax, xlabel='x', ylabel='y')
for i, color in enumerate('ryb'):
  mask = y == (i + 1)
  plt.scatter(X[mask, 0], X[mask, 1], c=color,
              label=f'{i+1}', cmap=plt.cm.RdYlBu, edgecolor='black')
```

### Generación de un árbol (elementos)

La construcción un árbol se realiza mediante un procedimiento recursivo en donde se aplica la función de corte $f_m(x)=x_i<=a$ 
 en el nodo 
, donde el parámetro $a$
 y la componente $x_i$
 se identifican utilizando los datos que llegan al nodo 
 de tal manera que se maximice una función de costo.
 
 Una función de costo podría estar basada en la entropía. Es decir para cada función de corte se mide la entroía en los nodos generados y se calcula la esperanza de la entropía de la siguiente manera: 
 
 $$L(x_i,a)$$

# Este código te proporciona la entropía asociada al conjunto de información



```python
def H(arr):
    a, b = np.unique(arr, return_counts=True)
    b = b / b.sum()
    return - (b * np.log2(b, where=b != 0)).sum()
```

### La función que optimiza L(xi,a)

Para encontrar el parámetro a, se utiliza la siguiente función de corte. Asume que las etiquetas están ordenadas por la variable x_i. la primera corresponde al valor mínimo y la última al valor máximo. El valor de $a$ es el índice con menor costo.  

```python
def corte_var(labels):
    mejor = (np.inf, None) # Guardar el valor de a con menor costo 
    D_m = labels.shape[0] # Se refiere al número de datos totales 
    corte = np.where(np.diff(labels))[0] + 1 # Se identifican los diferentes valores que se tienen que probar (aquellos puntos donde la clase cambia con respecto al elemento adyacente)
    # Dado que este quita el primer elemento, se debe incrementar uno.
    for j in corte: # El cíclo es por todos los puntos de corte, se calcula el costo para los elementos que están a la izquierda y derecha del corte y se compara el resultado
        # con el costo con menor valor encontrado hasta el momento
        izq = labels[:j]
        der = labels[j:] 
        a = (izq.shape[0] / D_m) * H(izq)
        b = (der.shape[0] / D_m) * H(der)
        perf = a + b
        if perf < mejor[0]:
            mejor = (perf, j)
    return mejor # revela el costo menor así como el índice donde se encontró.
```

Con el corte bar se optimiza el valor de $a$ de la función L(xi,a), ahora hay que optimizar xi con respecto a la función de costo. 

Para ello se utiliza el procedimiento corte, que encuentra el mínimo de xi, esta función recibe idx, donde se buscarán estos valores.



```python
# esta función busca el mínimo con respecto a x_i, recibe los índices idx, 
# donde se buscarán estos valores. En un inicio idx es de 0 al número de elementos
# del conjunto D menos uno.
def corte(idx):
    mejor = (np.inf, None, None) # define la variable donde se guardará el menor costo 
    orden = np.argsort(X[idx], axis=0) # se ordenan las variables
    labels = y[idx] # se obtienen las etiquetas involucradas
    for i, x in enumerate(orden.T):
        comp = corte_var(labels[x])
        if comp[0] < mejor[0]:
            mejor = (comp[0], i, comp[1])
    return mejor, idx[orden[:, mejor[1]]]
```

### Con respecto al proceso de entrenamiento la diferencia entre clasificación y regresión se encuentra en la función de costo que guía el proceso de optimización. En el caso de clasificación la función de costo era la esperanza de la entropía. Por otro lado, en regresión una función de costo utilizada es la varianza que es el error cuadrático que se muestra en los nodos. Para ejemplificar el uso de esta función de costo se utilizan los datos de Diabetes tal y como se muestran en las siguientes instrucciones.

```python
X, y = load_diabetes(return_X_y=True)
T, G, y_t, y_g = train_test_split(X, y, test_size=0.2)
```

Con los datos de entrenamiento se genera el siguiente árbol de decisión para regresión. Solamente se muestran la información de la raíz y sus dos hijos. En la raíz se observa los parámetros de la función de corte, se selecciona la variable con índice 8 y se envían 235 elementos al hijo izquierdo y el resto al hijo derecho.



```python
arbol = tree.DecisionTreeRegressor().fit(T, y_t)
_ = tree.plot_tree(arbol, max_depth=1)
```

El siguiente método implementa la función de corte para regresión se puede observar que la única diferente con la función corte_var definida en clasificación  es que la entropía H se cambia por la varianza np.var.

```python
def corte_var(response):
    mejor = (np.inf, None)
    D_m = response.shape[0]
    corte = np.where(np.diff(response))[0] + 1
    for j in corte:
        izq = response[:j]
        der = response[j:]
        a = (izq.shape[0] / D_m) * np.var(izq)
        b = (der.shape[0] / D_m) * np.var(der)
        perf = a + b
        if perf < mejor[0]:
          mejor = (perf, j)
    return mejor    
```

La función corte_var de regresión se utiliza para encontrar el punto de corte en los datos del conjunto de entrenamiento de la siguiente manera. En la primera línea se ordenan las variables independientes y en la segunda línea se itera por todas las variables independientes para calcular el corte con costo mínimo.

```python
orden = T.argsort(axis=0) # Ordena los puntos
res = [corte_var(y_t[orden[:, x]]) for x in range(10)]
res
```
